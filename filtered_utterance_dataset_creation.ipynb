```python
import json

# Load the JSON data from a file
file_path = 'data/news_dialogue_sample.json'
with open(file_path, 'r') as file:
    pre_data = json.load(file)

specific_id = "CNN-299342"
data = [next((item for item in pre_data if item['id'] == specific_id), None)]


# New list to hold the processed transcripts
processed_transcripts = []

# Function to process each transcript
def process_transcript(transcript):
    full_names = {}
    processed_entries = []
    
    # Process speaker names to find and replace abbreviations
    for i, speaker in enumerate(transcript['speaker']):
        if ',' in speaker:
            full_name = speaker.split(',')[0]
            last_name = full_name.split()[-1]
            full_names[last_name] = full_name
            speaker = full_name
        else:
            speaker = full_names.get(speaker, speaker)
        
        # Create a new dictionary for each transcript row
        entry = {
            'transcript_id': transcript['id'],
            'program': transcript['program'],
            'date': transcript['date'],
            'url': transcript['url'],
            'summary': transcript['summary'],
            'transcript_row_num': i,
            'speaker': speaker,
            'utt': transcript['utt'][i]
        }
        processed_entries.append(entry)
    
    return processed_entries

# Process each transcript in the dataset
for transcript in data:
    processed = process_transcript(transcript)
    processed_transcripts.extend(processed)

# Optionally, write the processed data to a new JSON file
output_path = 'data/processed_news_dialogue_sample.json'
with open(output_path, 'w') as outfile:
    json.dump(processed_transcripts, outfile, indent=4)

# Print a sample for verification
# print(json.dumps(processed_transcripts, indent=4))  # Print first two entries for quick check
```

## Make the filtered_utterances dataset

```python
import json
import re

# Load the transcripts and phrases from JSON files
with open('data/news_dialogue_large_sample.json', 'r') as file:
    transcripts = json.load(file)

with open('data/short_prm_list.json', 'r') as file:
    pprm_list = json.load(file)

# Create a list of words/phrases to search for
search_terms = pprm_list

def filter_and_prefix_utterances(transcripts, terms, max = 1000):
    # Function to filter utterances, collect matching terms, and include the previous utterance
    results = []
    for transcript in transcripts:
        previous_utterance = None  # Initialize the previous utterance tracker
        for current_index, utterance in enumerate(transcript['utt']):
            matched_terms = {}  # Change to an empty dictionary
            modified_utterance = utterance
            # Check each term separately
            for term in terms:
                # Use regex to match whole words only
                if re.search(rf"\b{re.escape(term)}\b", utterance, re.IGNORECASE):
                    # Add the term to the dictionary with value "ungraded"
                    matched_terms[term] = "ungraded"
                    # Use regex to replace only the first case variation of the term
                    modified_utterance = re.sub(rf"\b{re.escape(term)}\b", f"<{term.upper()}>", modified_utterance, count=1, flags=re.IGNORECASE)
            # If any terms matched, collect the previous and current utterance
            if matched_terms:
                result = {
                    "transcript_id": transcript['id'],
                    "matched_terms": matched_terms,
                    "previous_statement": previous_utterance if previous_utterance else "None",
                    "statement": modified_utterance
                }
                results.append(result)
                # Stop collecting after 10 entries to manage output size
                
            # Update the previous utterance
            previous_utterance = utterance
    return results


# Get the filtered list of utterances with prefixed terms and previous context
filtered_utterances = filter_and_prefix_utterances(transcripts, search_terms)

import random

# Sample 1000 entries randomly from the results if there are more than 1000
sampled_utterances = random.sample(filtered_utterances, min(100, len(filtered_utterances)))

# Save the filtered utterances to a JSON file
output_file_path = 'data/filtered_utterances.json'
with open(output_file_path, 'w') as file:
    json.dump(filtered_utterances, file, indent=4)

output_file_path = 'data/filtered_utterances_sample.json'
with open(output_file_path, 'w') as file:
    json.dump(sampled_utterances, file, indent=4)

# Print the JSON in a readable way
# print(json.dumps(sampled_utterances, indent=4))
```

```python
import json

# Sample JSON data
with open('data/new_filtered_utterances_to_grade/filtered_data_MW_part_3.json', 'r') as prev_grade:
    prev_grade1 = json.load(prev_grade)
with open('data/new_filtered_utterances_to_grade/filtered_data_MW_part_2.json', 'r') as prev_grade:
    prev_grade2 = prev_grade1+json.load(prev_grade)

# Dictionary to store the count of each term for each category
term_count = {}

# Iterate through each item in the data
for item in prev_grade2:
    matched_terms = item['matched_terms']
    for term, category in matched_terms.items():
        # Initialize the term dictionary if not already present
        if term not in term_count:
            term_count[term] = {}
        # Initialize the category count if not already present
        if category not in term_count[term]:
            term_count[term][category] = 0
        # Increment the count for this category under the term
        term_count[term][category] += 1


sorted_term_count = dict(sorted(term_count.items(), key=lambda item: item[1].get('none', 0), reverse=True))

# Output the term count dictionary
# print(json.dumps(term_count, indent=2))
for item in sorted_term_count:
    # print(item, term_count[item])
    values = term_count[item]
    sample_count = 0
    none_count = 0
    for k, v in values.items():
        sample_count+=v
        if k == "none":
            none_count += v
    print(f"{item}: none percentage: {none_count/sample_count*100:.2f}% \n * h/a_count: {sample_count-none_count}\n * none_count: {none_count}\n")
```
